```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36 Edg/136.0.0.0',
    'Cookie': 'bid=UMtW4Kj7qmw; _pk_id.100001.4cf6=74169311fb5d6b90.1746692425.; _pk_ses.100001.4cf6=1; __utma=30149280.214996804.1746692425.1746692425.1746692425.1; __utmb=30149280.0.10.1746692425; __utmc=30149280; __utmz=30149280.1746692425.1.1.utmcsr=(direct)|utmccn=(direct)|utmcmd=(none); __utma=223695111.1269508625.1746692425.1746692425.1746692425.1; __utmb=223695111.0.10.1746692425; __utmc=223695111; __utmz=223695111.1746692425.1.1.utmcsr=(direct)|utmccn=(direct)|utmcmd=(none); ap_v=0,6.0; dbcl2="287656738:Yi6HbeV04GE"; ck=m9GE; push_noty_num=0; push_doumail_num=0',
}

url = 'https://movie.douban.com/subject/34780991/reviews'
print('\n-----------获取的页面-----------\n', end='\n\n')
numbers = 1
all_comments_data = []


# 遍历多页评论
for i in range(0, 7841, 20):  # 每页有20条评论
    pages = i // 20
    last_pages = 7840 // 20 - pages
    i = str(i)
    urls = url + '?sort=hotest&start=' + i
    print(f"------------- 获取第{pages + 1}页评论 --------------")
    try:
        response = requests.get(urls, headers=headers)
        if response.status_code == 200:
            # 使用 BeautifulSoup 解析 HTML
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 查找所有 class='review-list' 的 div 元素
            reviewlists = soup.find_all('div', class_='review-list')
            
            for reviewlist in reviewlists:
                # 找到所有包含 data-cid 的 div 元素
                review_items = reviewlist.find_all('div', attrs={'data-cid': True})
                for review_item in review_items:
                    # 找到 main review-item 下的 main-hd
                    main_hd = review_item.find('header', class_='main-hd')
                    if main_hd:
                        # 提取 a:name
                        name = main_hd.find('a', class_='name').get_text(strip=True)
                        
                        # 提取第一个 span 的 title 属性和第二个 span 的文本内容
                        first_span = main_hd.find('span')  # 找到第一个 span
                        first_span_title = first_span.get('title', '') if first_span else ''  # 获取 title 属性
                        second_span = main_hd.find_all('span')[1] if len(main_hd.find_all('span')) > 1 else None
                        second_span_text = second_span.get_text(strip=True) if second_span else ''
                        
                        # 找到 main review-item 下的 main-bd
                        main_bd = review_item.find('div', class_='main-bd')
                        if main_bd:
                            # 提取 h2 下的 a 链接文字内容
                            h2_a_text = main_bd.find('h2').find('a').get_text(strip=True)
                            
                            # 将提取的数据存储到全局列表中
                            all_comments_data.append({
                                '序号': len(all_comments_data) + 1,
                                '评论用户': name,
                                '评分': first_span_title,
                                '评论时间': second_span_text,
                                '精炼评论': h2_a_text
                            })
        else:
            print(f"请求失败，状态码：{response.status_code}")
    except Exception as e:
        print(f"在处理第{pages + 1}页时发生错误：{e}")

    print(f"------------- 第{pages + 1}页评论已经获取，余下{last_pages}页 --------------")
    print(end='\n\n')
    # 随机暂停一段时间，避免触发人机验证
    time.sleep(random.randint(1, 3))
```

```python
# 将所有数据转换为 DataFrame
df = pd.DataFrame(all_comments_data)

# 保存为 Excel 文件
df.to_excel('comments.xlsx', index=False)
print("所有数据已成功保存到 comments.xlsx 文件中。")

# 保存为 Json 文件
df.to_json('comments.json')
print("所有数据已成功保存到 comments.json 文件中。")
```
